```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split

from PIL import Image
```

```
img = Image.open('/content/drive/My Drive/strange.jpg')
```

```
small = img.resize(size=(400,300))
```

```
small
```

```
img_arr = np.array(small)
```

```
plt.imshow(img_arr)
```

```
img_arr.shape
```

```
X = img_arr.reshape(-1,3)
```

```
X.shape
```

```
from sklearn.cluster import KMeans
```

```
model = KMeans(n_clusters=10)
```

```
model.fit(X)
```

```
model.cluster_centers_ # here range of RGB should be in range(255) so changing its astype 
```

```
pallets = model.cluster_centers_.astype(np.uint8)
```

```
pallets
```

```
# converting pallets into 3D
```

```
plt.imshow(pallets.reshape(1,10,3)) # this color used to generate image
```

```
model.labels_[:100] # pixel colors of each labels
```

```
output = pallets[model.labels_]
```

```
output[:100] # it show the all pixels colors
```

```
output.shape
```

```
out_img = output.reshape(img_arr.shape) # reshaping into original data shape
```

```
 plt.imshow(out_img) # this image is made by 6 colors as 6 cluster center we used
```
